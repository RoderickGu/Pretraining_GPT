```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/FullData/full_dialog.csv").iloc[:, 1:]
```

```python
dialog_list = set(item for item in set(df["B2"]) if item[:3] != "BAD")
all_dialogs = []

for tag in dialog_list:
    all_dialogs.append(df[df["B2"] == tag])
```

```python
indices = np.arange(len(all_dialogs))
np.random.shuffle(indices)
cut_index = int(len(all_dialogs)*0.9)
train_indices = indices[:cut_index]
val_indices = indices[cut_index:]
```

```python
raw_train_dialogs = [all_dialogs[i] for i in train_indices]
raw_val_dialogs = [all_dialogs[i] for i in val_indices]
```

```python
torch.save(raw_train_dialogs, "raw_train_dialogs.pkl")
torch.save(raw_val_dialogs, "raw_val_dialogs.pkl")
```
